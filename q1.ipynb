```python
import matplotlib.pyplot as plt
from math import cos, sin, pi
import numpy as np

x = []
y = []
n=256
for theta in np.linspace(0,n*pi):
    r = ((theta)**2)
    radius = np.linspace(.5,1.,n)
    x.append(r*cos(theta))
    y.append(r*sin(theta))

plt.figure()

plt.scatter(x,y)  
plt.show()
```
