# Practicing neural network activation functions


```
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,11,1)
print(x)

def plot_graph(u,ylabel):
  plt.figure()
  plt.plot(x,y,'-o')
  plt.grid(True)
  plt.xlabel("X")
  plt.ylabel(ylabel)
  plt.show()

# Step Function
y = list(map(lambda a: 1 if a>=0 else 0,x))
# Sigmoid function
plot_graph(y,"STEP(X")
```

```
y = 1 / (1+np.exp(-x))
plot_graph(y,"Sigmoid(X)")
```

```
# Tanh
y = (np.exp(2*x)-1) / (np.exp(2*x) + 1)
plot_graph(y,"TanH(X)")
```

```
# ReLU
y = list(map(lambda a:a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

```
# Leaky ReLU

alpha = 0.1

y = list(map(lambda a:a if a>=0 else alpha*a,x))
plot_graph(y,"Leaky Re-LU")
```

```
# Softmax

arr = [6, 2, 9]

p1 = np.exp(arr[0])/(np.exp(arr[0])+np.exp(arr[1])+np.exp(arr[2]))
p2 = np.exp(arr[1])/(np.exp(arr[0])+np.exp(arr[1])+np.exp(arr[2]))
p3 = np.exp(arr[2])/(np.exp(arr[0])+np.exp(arr[1])+np.exp(arr[2]))

print(p1,p2,p3)
print(p1+p2+p3)

data = [5, 8, 4, 11, 3, 12]
#plt.bar(range(len(data)),data)


def softmax(arr):
  e = np.exp(arr)
  return e/e.sum()

res = softmax(data)
print(res)
print(np.sum(res))


plt.bar(range(len(res)),res)
```
